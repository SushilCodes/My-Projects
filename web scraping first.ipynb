```python
import requests
import bs4

from bs4 import BeautifulSoup
```

```python
url='https://timesofindia.indiatimes.com/sports/cricket/india-in-england'
```

```python
headers={                               #use in console type navigation.useragent
    'User-Agent: "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari" '
}
```

```python
r=requests.get(url,{'headers':headers})
```

```python
soup=bs4.BeautifulSoup(r.text,'html.parser')
```

```python
soup.find_all() #to find all source code
```

```python
soup.find_all('div') #to find div
```

```python
#html tree traversal

#we can use commonly four soup objects
#tag,Navigation String,Beautisoup,and comment

#get the title of web page

title=soup.title
```
